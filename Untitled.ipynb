```python
import pandas as pd
import pyarrow as pa
import fastparquet as fp

# Reading a Parquet file
dataHTTP = pd.read_parquet('Testing.parquet')

dataHTTP
```

```python
dataHTTP.size
```

```python
from pycaret.clustering import *

import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

model = ClusteringExperiment()
model.setup(data = dataHTTP,
            normalize = True,
            #categorical_features = ['status'],
            numeric_features = ["domain_registration_length","length_url","domain_age","web_traffic","dns_record"])
```

```python
kmeans = model.create_model('kmeans', num_clusters = 2)
```

```python
model.evaluate_model(kmeans)
```

```python
model.plot_model(kmeans, plot = 'silhouette')
```

```python
result = model.assign_model(kmeans)
result.head()
```

```python
predictions = model.predict_model(kmeans, data = dataHTTP)
predictions.head()
```
